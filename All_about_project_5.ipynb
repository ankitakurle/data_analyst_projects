# 2) CrimeTime

```python
data['CrimeTime']
```

```python
# we can see that crime time is not in right format
data.CrimeTime.unique()
```

```python
# defining a function to convert the time in correct format.
# F-strings make the code more readable by allowing us to embed variables directly within the string, eliminating the need for concatenation
# using replace method to remove the colon and finding whether the time contains 6 digits or not if not we are filling it with zfill
def standardize_time(t):
    t = t.replace(':', '').zfill(6)
    return f"{t[:2]}:{t[2:4]}:{t[4:6]}"
```

```python
# applying function on time
data['standardized_time'] = data['CrimeTime'].apply(standardize_time)
data
```

```python
data.standardized_time.value_counts()
```

```python
data.District
```

```python
# changing datatype from object to string
data.District=data.District.astype(str)
```

```python
# finding the unique regions
data.District.unique()
```

```python
# formatting regions
def format_region(regi):
    regi=regi.replace('NORTHEASTERN','NORTH-EASTERN').replace('SOUTHEASTERN','SOUTH-EASTERN').replace('SOUTHWESTERN','SOUTH-WESTERN').replace('NORTHWESTERN','NORTH-WESTERN')
    regi=regi.title()
    return regi
```

```python
# appling the correct formate of regions
data.District= data.District.apply(format_region)
```
